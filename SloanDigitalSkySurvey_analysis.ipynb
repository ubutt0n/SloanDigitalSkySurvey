```python
#Удалим все столбцы оканчивающиеся на id
```

```python
dataset = dataset.drop(["objid"], axis=1)
```

```python
dataset = dataset.drop(["specobjid"], axis=1)
```

```python
dataset = dataset.drop(["fiberid"], axis=1)
```

```python
#Также удалим столбец rerun, т.к. он имеет только одно значение
```

```python
dataset = dataset.drop(["rerun"], axis=1)
```

```python
#В столбце class заменим названия классов на цифры
```

```python
class_set = set(dataset["class"])
i = 0
for item in class_set:
    dataset["class"] = dataset["class"].replace(item, i)
    i += 1
```

```python
#Проведем нормализацию столбоцов с числовыми значениями
```

```python
import math
```

```python
numerical = ["ra", "dec", "u", "g", "r", "i", "z", "redshift", "mjd"]

for column in numerical:
    dataset[column] = (dataset[column] - dataset[column].mean()) / (math.sqrt(dataset[column].var()))
```

```python
arr = []
for column in numerical:
    arr.append(dataset[column].mean())
arr
```

```python
#Наибольшая корреляция наблюдается между r и g, i и g, r и i, r и z, plate и mjd
```

```python
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
standart_X_train = pd.DataFrame(preprocessing.scale(X_train), columns=X_train.columns)
```

```python
for i in range(1, len(standart_X_train.columns)):
    pca = PCA(n_components=i)
    pca.fit(standart_X_train)
    print(i, sum(pca.explained_variance_ratio_))
```

```python
#7 признаков достаточно для объяснения 90% дисперсии
```

```python
pca = PCA(n_components=7)
pca.fit(standart_X_train)
```

```python
greatest_contributor = pca.components_[0][0]
gr_contrib_i = 0
for i in range(0, len(pca.components_[0])):
    if abs(pca.components_[0][i]) > abs(greatest_contributor):
        greatest_contributor = pca.components_[0][i]
        gr_contrib_i = i
print(gr_contrib_i, greatest_contributor)
```

```python
#Наибольший вклад в первую компоненту вносит признак r
```

```python
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=7)
pca.fit(standart_X_train)
```

```python
pca_x = pca.transform(standart_X_train)
```

```python
tsne = TSNE(n_components=2)
tsne_x = tsne.fit_transform(pca_x)
```

```python
tsne_dataframe = pd.DataFrame(tsne_x, columns=["X", "Y"])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(tsne_dataframe["X"], tsne_dataframe["Y"], c=y_train, s=10)
```
